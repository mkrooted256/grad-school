```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import sympy as sym
```

```python
def f(x, a=10.0, b=10.0):
    x = x.round(3)
    y = x**2
    return y.round(3)

def sf(x, a=10.0):
    return x**2
```

```python
X = np.linspace(-0.1,0.1,200)
plt.plot(X, f(X))
```

```python
SX = np.array(list(set(X.astype(np.float16))))
SX.sort()
plt.plot(SX, f(SX))
```
