Нехай $n > 2$.
Розглянемо 2 випадки.

1. Нехай $n=p$ - просте.  
    Тоді за Властивістю 3.3 в $Z/nZ$ існуватиме обернений до $a$ елемент.
2. Нехай $n$ - не просте.  
    Тоді $n = \prod_{p\in\mathcal P} p^{n_p}$.   
    1. Припустимо, що $\gcd(a,n) = 1$.
        Тоді
        $$\exists x,y\in\mathbb Z: \; ax+ny=1, \; (x\ne 0, y\ne 0)$$
        Візьмемо обидві частини р-ня за модулем $n$
        $$ax = 1 \mod n$$
        Отже, існує обернений елемент.
        
    2. Інакше $\gcd(a,n) = g$.  
        Тоді $a = ug, \; n = vg, \; \gcd(u,v)=1$.  
        Зауважу, що $|n|=|v||g|$, тобто $|v|<n$.  
        Тоді 
        $$a \cdot v = (ug)v = u(vg) = un = 0 \mod n$$
        Тобто знайшли таке $v\in Z/nZ$ що $av = 0 \mod n$

### Ex 3.4.1
$$m = \sum_{k\ge 0} m_k \cdot 10^k$$  

$$m = \sum_{k\ge 0} m_k \cdot 10^k \mod 9$$
$$m = \sum_{k\ge 0} m_k \cdot 1^k = \sum_{k\ge 0} m_k\mod 9$$
Тобто $9\mid m \iff m = 0 \mod 9 \iff \sum_{k\ge 0}m_k = 0 \mod 9 \iff 9\mid \sum_{k\ge 0}m_k $

### Ex 3.4.2
$$m = \sum_{k\ge 0} m_k \cdot 10^k$$  
$$m' = \sum_{k\ge 0} m_{\pi(k)} \cdot 10^k, \quad \pi \text{ - перестановка}$$  

$$(m-m') = \sum_{k\ge 0} (m_k-m_{\pi(k)}) \cdot 10^k = \sum_{k\ge 0} (m_k-m_{\pi(k)}) \cdot 1^k = \sum_{k\ge 0} (m_k-m_{\pi(k)}) = \sum_{k\ge 0} m_k - \sum_{k\ge 0} m_{\pi(k)} = 0\mod 9$$
Тобто $9\mid(m-m')$

## Ex 3.6
Представити число 441 в системі числення з базою 13.

```python
N = 441
b = 13
all_digits = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'

n = N
digits = []
step = 1
while n > 0:
    d = n % b   # modulo
    n = n // b  # integer division
    digits.append(d)
    print(f"step {step}: n={n}, d={d}")
digits.reverse()

print(f"({N})_10 is ({''.join([all_digits[d] for d in digits])})_{b} or {digits}")

m = 0
for k,a_k in enumerate(digits):
    m *= b
    m += a_k
print(f"Checking: {N} == {m}")
```

Отже, 
$$441 = 2\cdot13^2 + 7\cdot 13 + 12$$

$$[a]_n[b]_n = [ab]_n$$
